# Data Balancing

## Using SMOTE

```python
from imblearn.over_sampling import SMOTE
import pandas as pd
from pathlib import Path

df = pd.read_csv(Path("dataset/preprocessed/train.csv"))
df = df[df["label"] != "~"]

labels = df["label"].tolist()
# Assuming 'record_name' corresponds to the file names of the signals
signal_data = []
for record in df["record_name"]:
    signal_path = Path(f"dataset/preprocessed/train/{record}.csv")
    signal_df = pd.read_csv(signal_path, header=None, index_col=None).to_numpy().flatten()
    signal_data.append(signal_df)
```

```python
smote = SMOTE()
signal_data_resampled, labels_resampled = smote.fit_resample(signal_data, labels)
```

```python
labels_df = pd.DataFrame(labels_resampled, columns=["label"])
labels_df.value_counts()
```

```python
import neurokit2 as nk
ind = 0
print(labels_resampled[len(labels)+ind])
nk.signal_plot(signal_data_resampled[len(signal_data)+ind], figsize=(24, 6))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(30, 6))
s1, s2 = 1, 68
plt.plot(signal_data[s1], label=labels[s1])
plt.plot(signal_data[s2], label=labels[s2])
plt.legend()
plt.show()
```
